```python
def hebin(_s_slot, _o_slot):
    s_slot = copy.deepcopy(_s_slot)
    o_slot = copy.deepcopy(_o_slot)
    res = {}
    for k, v in s_slot.items():
        if k in o_slot:
            if isinstance(v, str) and isinstance(o_slot[k], list):
                res[k] = v
            elif isinstance(v, list) and isinstance(o_slot[k], str):
                res[k] = o_slot[k]
            elif isinstance(v, str) and isinstance(o_slot[k], str):
                if v in o_slot[k]:
                    res[k] = o_slot[k]
                elif o_slot[k] in v:
                    res[k] = v
#                 elif len(o_slot[k]) >= len(v):
#                     res[k] = o_slot[k]
                else:
                    res[k] = o_slot[k]
        else:
            res[k] = v
            
    for k, v in o_slot.items():
        if k not in res:
            res[k] = v
            
    new_s_slot = copy.deepcopy(_s_slot)
    new_o_slot = copy.deepcopy(_o_slot)
    all_value = []
    origin_res = copy.deepcopy(res)
    rm_k = []
    for k, v in res.items():
        if isinstance(v, list):
            continue
        for ok,ov in origin_res.items():
            if isinstance(ov, list):
                continue
            if ok != k and (v in ov or ov in v):
                if k in new_o_slot:
                    pass
                else:
                    rm_k.append(k)
    rm_k = list(set(rm_k))
#     print(rm_k)
    res = {k:v for k, v in res.items() if k not in rm_k}
    return res
```
