```python
def hebin(s_slot, o_slot):
    res = {}
    for k, v in s_slot.items():
        if k in o_slot:
            if isinstance(v, list):
                res[k] = v
            else:
                res[k] = o_slot[k]
        else:
            res[k] = v
            
    for k, v in o_slot.items():
        if k not in res:
            res[k] = v
            
    return res
```
