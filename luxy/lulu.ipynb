```python
t = '付 刘永义 车款：385000元（主车385000元，挂车0元）,有定金：10000元）,有代收款：0元，实付：375000元,费用明细：(首付款0元,GPS费:1000元,保险费(交强险):5000元,保险费(商业险):15000元,服务费:18100元,杂费:2000元,落户保证金:2000元,还款保证金:5000元 合计48100元).'
t
```

```python

# 抽取 [首付款|GPS费|保险费(交强险)|保险费(商业险)|杂费|落户保证金|还款保证金|落户费|附加税|服务费|主车|挂车|有定金|实付]
def _search(t):
    res = {
        '首付款': '',
        'GPS费': '',
        '保险费(交强险)': '',
        '保险费(商业险)': '',
        '杂费': '',
        '落户保证金': '',
        '还款保证金': '',
        '落户费': '',
        '附加税': '',
        '服务费': '',
        '主车': '',
        '挂车': '',
        '有定金': '',
        '实付': ''
    }
    # 首付款
    sfk = re.compile('首付款.{0,3}?([0-9]+)')
    r_sfk = re.search(sfk, t)
    if r_sfk:
        res['首付款'] = r_sfk.groups()[-1]
    # GPS费
    gps = re.compile('GPS费.{0,3}?([0-9]+)')
    r_gps = re.search(gps, t)
    if r_gps:
        res['GPS费'] = r_gps.groups()[-1]
    # 保险费(交强险)
    bxf = re.compile('保险费.{0,1}交强险.{0,3}?([0-9]+)')
    r_bxf = re.search(bxf, t)
    if r_bxf:
        res['保险费(交强险)'] = r_bxf.groups()[-1]
    # 保险费(商业险)
    bxf_sy = re.compile('保险费.{0,1}商业险.{0,3}([0-9]+)')
    r_bxf_sy = re.search(bxf_sy, t)
    if r_bxf_sy:
        res['保险费(商业险)'] = r_bxf_sy.groups()[-1]
    # 杂费
    zf = re.compile('杂费.{0,3}?([0-9]+)')
    r_zf = re.search(zf, t)
    if r_zf:
        res['杂费'] = r_zf.groups()[-1]
    # 落户保证金
    
    print(res)

def sub_search(p, t):
    pattern = re.compile(p+'.{0,3}?([0-9]+)')
    res = re.search(pattern, t)
    if res:
        return res.groups()[-1]
    return ''

def all_search(t):
    res = {
        '首付款': '',
        'GPS费': '',
        '保险费(交强险)': '',
        '保险费(商业险)': '',
        '杂费': '',
        '落户保证金': '',
        '还款保证金': '',
        '落户费': '',
        '附加税': '',
        '服务费': '',
        '主车': '',
        '挂车': '',
        '有定金': '',
        '实付': ''
    }
    for k, v in res.items():
        if '(' in k:
            p = k.replace('(', '.{0,1}')
            p = p.replace(')', '')
#             print(p)
        else:
            p = k
        res[k] = sub_search(p, t)
    return res

print(all_search(t))
```

```python
def _new(x, df, col_name, last=2):
    int_last = int(x[-last:])
    for i in range(int_last, 10**last):
        if x[:-last]+str(i) not in df[col_name].values:
#             print(x[:-last]+str(i))
            return x[:-last]+str(i)
    print('-----')
```
